```python
from datasets import load_from_disk

dataset=load_from_disk(r"orgin_datasets")
```

```python
from dotenv import load_dotenv
load_dotenv()
```

Deepseek

```python
from openai import OpenAI
import os
deepseek_api_key =os.getenv("deepseek_api")
client_deepseek = OpenAI(api_key=deepseek_api_key, base_url="https://api.deepseek.com")

def deepseek(text):
      response = client_deepseek.chat.completions.create(
          model="deepseek-chat",
          messages=[
            {"role":"system","content":"You are a code generation assistant,Use a function to implement the following code"},
            {"role":"user","content":text},
          ],
          stream=False
      )
      return response.choices[0].message.content
```

```python
def process_prompt(example):
    prompt=example["prompt"]
    model_answer=deepseek(prompt)
    # print(model_answer)
    return{'model_answer':model_answer}

dataset['train']=dataset['train'].map(process_prompt)
```

```python
dataset.save_to_disk("deepseek_datasets")
```
